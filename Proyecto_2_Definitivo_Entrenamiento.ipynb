### Contexto  
Mejorar la satisfacción de los pasajeros es crucial por varias razones. 

- En primer lugar, es más probable que los pasajeros satisfechos se conviertan en clientes fieles y recomienden los servicios a otras personas, lo que conduce a una mayor retención de clientes y a un marketing boca a boca positivo.

- En segundo lugar, el análisis de los datos de satisfacción de los pasajeros permite a las organizaciones identificar los puntos débiles y las áreas de mejora, lo que les permite tomar decisiones basadas en datos y mejorar la experiencia general de los pasajeros. 

Al aprovechar los datos para comprender y atender las necesidades de los pasajeros, las empresas pueden esforzarse por mejorar continuamente y mantener una ventaja competitiva en el sector.

```python
def select_columns(df: pd.DataFrame, threshold: int = 10) -> list[str]:
    """
  Función utilizada para dividir las columnas de un DataFrame en discretas y continuas en función del umbral especificado.

    Args:
        df (pd.DataFrame): DataFrame que contiene las columnas a seleccionar.
        threshold (int): Umbral que define el número mínimo de valores únicos. Por defecto es 10.

    Devuelve:
        columnas_discretas (lista[str]): Lista de nombres de columnas discretas.
        continuous_columns (lista[str]): Lista de nombres de columnas continuas.
    """
    
    # Inicializamos una lista de columnas discretas
    discrete_columns = []

    for column in df.columns:
        # Seleccionar las columnas con un número de valores únicos inferior o igual al umbral
        if df[column].nunique() <= threshold:
            # Añade las columnas seleccionadas a la lista de columnas discretas
            discrete_columns.append(column)
    
    # Selecionamos las columnas continuas.
    continuous_columns = [col for col in df.columns.tolist() if col not in discrete_columns]


    return discrete_columns, continuous_columns
```

```python
# Aplicamos la función select_columns
discrete_columns, continuous_columns = select_columns(train_df)
```

```python
def visualizations_discrete(df: pd.DataFrame):
    """
    Función utilizada para trazar gráficos de recuento para cada columna discreta.

    Parámetros:
        - df (pd.DataFrame): DataFrame que contiene los datos.
    """

    # Establecemos el tamaño y diseño de los gráficos
    fig, axes = plt.subplots(nrows=len(discrete_columns), ncols=2, figsize=(12, 6 * len(discrete_columns)))
    fig.subplots_adjust(hspace=0.25)
    sns.color_palette("Blues", as_cmap=True)
    
    for index, column in enumerate(discrete_columns):
        # Trazamos countplot de cada columna
        sns.countplot(data=df, x=column, ax=axes[index, 0], palette="blend:#7AB,#EDA")
        axes[index, 0].set_title(f"Countplot of {column}")
        axes[index, 0].set_xlabel("")

        # Trazamos countplot de cada columna con respecto al Target (Satisfacción)
        sns.countplot(data=df, x=column, hue=TARGET, ax=axes[index, 1], palette="blend:#7AB,#EDA")
        axes[index, 1].set_title(f"Countplot of {column} specified by {TARGET}")
        axes[index, 1].set_xlabel("")  

    # Mostramos los gáficos
    plt.show()
```

```python
def visualizations_continuous(df: pd.DataFrame, target: str):
    """
    Función utilizada para trazar el boxplot y la distribución de cada columna continua.

    Parámetros:
        - df (pd.DataFrame): DataFrame que contiene los datos.
        - target (str): Variable objetivo de un DataFrame.
    """
    
    # Establecemos el tamaño y diseño de los gráficos
    fig, axes = plt.subplots(nrows=len(continuous_columns), ncols=2, figsize=(12, 6 * len(continuous_columns)))
    fig.tight_layout(pad=2.0)

    for index, column in enumerate(continuous_columns):
        # Trazamos boxplot
        sns.boxplot(data=df, x=TARGET, y=column, ax=axes[index, 0], palette="blend:#7AB,#EDA")
        axes[index, 0].set_title(f"Boxplot of {column}")
        axes[index, 0].set_xlabel("")

        # Trazamos distribuciones
        sns.histplot(data=df, x=column, kde=True, ax=axes[index, 1], palette="blend:#7AB,#EDA")
        axes[index, 1].set_title(f"Distribution of {column}")
        axes[index, 1].set_xlabel("")
        
    # Mostramos los gáficos
    plt.show()
```

```python
# Matriz de modelos que vamos a estudiar
modelos = [
          # KNeighborsClassifier(),
           LogisticRegression(),
           DecisionTreeClassifier(),
           RandomForestClassifier(),
           # SVC(),
           #AdaBoostClassifier(),
           #GradientBoostingClassifier(),
          ]
```

```python
# Pasamos las variavles en los modelos y obtenemos los balores de la metricas 


datos_modelos = list()

for model in modelos:
    
    model.fit(X_train, y_train)
    yhat = model.predict(X_test)
    
    
    # Metricas
    acc = accuracy_score(y_test, yhat)
    jac = jaccard_score(y_test, yhat)
    pre = precision_score(y_test, yhat)
    rec = recall_score(y_test, yhat)
    f1s = f1_score(y_test, yhat)
    roc = roc_auc_score(y_test, yhat)
    cmt = confusion_matrix(y_test, yhat)
    
    
    
    datos_modelos.append([str(model), model, acc, jac, pre, rec, f1s, roc, cmt])
    
df_modelo = pd.DataFrame(data = datos_modelos, columns = ["name", "model", "acc", "jac", "pre", "rec", "f1s", "roc", "cmt"])

df_modelo.sort_values("roc", ascending = False)
```
